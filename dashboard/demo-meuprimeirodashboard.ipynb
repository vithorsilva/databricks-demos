# Roteiro

## Criando um Dashboard

Conecte-se ao seu ambiente Databricks (inclusive o Free Edition), clique em Dashboards e depois em "Create Dashboard".
![dashboard1-tutorial-1](./images/dashboard1-tutorial-1.png "Tutorial - Meu Primeiro Dashboard")

Uma nova tela será aberta, destacando pontos importantes:
![dashboard1-tutorial-1](./images/dashboard1-tutorial-2.png "Tutorial - Meu Primeiro Dashboard")

* **Data**: Adicione os dados que serão usados nos itens visuais (gráficos). Os Dashboards do Databricks utilizam datasets, permitindo que cada gráfico esteja vinculado a um dataset específico. Para relacioná-los, utilize filtros ou parâmetros.
* **Páginas**: É possível criar dashboards com várias abas/páginas, permitindo interação entre os dados. Distribua as informações para facilitar a navegação.
* **Filtros Globais**: Filtros impactam todos os gráficos/datasets vinculados, tornando a análise dinâmica.
* **Visualizações**: Adicione diferentes tipos de gráficos (barras, linhas, pizza, tabelas, mapas, etc.), personalize cores, títulos, eixos e legendas.
* **Parâmetros**: Crie parâmetros dinâmicos para personalizar consultas SQL ou visualizações, tornando o dashboard interativo.
* **Layout**: Organize os elementos do dashboard arrastando e redimensionando itens, criando uma experiência visual adequada.
* **Compartilhamento**: Compartilhe dashboards com outros usuários, defina como público ou privado e agende atualizações automáticas.
* **Exportação**: Exporte dashboards em PDF ou imagens para apresentações externas.
* **Comentários**: Adicione comentários e anotações diretamente nos dashboards para facilitar a colaboração.
* **Agendamento**: Configure atualizações automáticas dos dados para manter as visualizações sempre atualizadas.

Essas funcionalidades permitem criar dashboards completos, interativos e colaborativos no Databricks, facilitando a análise e visualização dos dados.

## Passo a passo

1. Nomeie o dashboard como "dash-vendas-pipeline"
2. Nome a aba para "Indicadores"
3. Clique em "Data" para adicionarmos uma instrução SQL, ao qual será:

```sql
select
    -- Dados sobre o Lead (Dimensões)
    o.opportunityid, o.type, o.name as opportunity, /*o.description,*/ o.createddate, o.closedate, o.days_to_close, o.leadsource,
    -- Dados sobre a Conta (Dimensões)
    o.accountid, o.account_name, o.industry as account_industry, o.business_type__c as industry_businesstype, 
    o.region as account_region, o.company_size_segment__c as account_sizesegment, o.region_size as account_region_size, o.annualrevenue as account_annualrevenue,
    -- Dados sobre o Vendedor (Dimensões)
    o.ownerid, u.name as salesman, sup.name as salesman_manager ,
    o.probability,
    o.stagename,
    o.forecastcategory,
    -- Medidas
    o.amount, o.opportunity_amount,
    o.New_Expansion_Booking_Annual__c,
    o.New_Expansion_Booking_Annual_Wtd__c,
    o.New_Recurring_Bookings_Manual__c,
    o.New_Recurring_Bookings__c
from main.dbdemos_aibi_sales_pipeline_review.opportunity o
inner join main.dbdemos_aibi_sales_pipeline_review.user u on o.ownerid = u.id
LEFT join main.dbdemos_aibi_sales_pipeline_review.user sup on sup.id = u.managerid
```
